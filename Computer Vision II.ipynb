## Open CV - Open Source Computer Vision
- Contains many popular algorithm that is being used for computer vision, object detection and tracking.

```python
# Install Open CV
!pip install opencv-python
```

```python
import cv2
```

```python
# Use OpenCV to load the image
buhari = cv2.imread(r'C:\Users\user\OneDrive\Desktop\buhari.jpg')
type(buhari)
```

```python
# Dimensionality
buhari.shape
```

```python
# Plot the image -  OpenCV uses BGR (Blue, Green and Red) instead of RGB (Red, Green and Blue)
plt.imshow(buhari)
```

```python
# Convert the image from BGR to RGB
bubu = cv2.cvtColor(buhari, cv2.COLOR_BGR2RGB)
plt.imshow(bubu)
```

```python
# Resizing the image - (Width, Height)
plt.imshow(cv2.resize(bubu, (600, 600)))
```

```python
# Assign resized image to a new variable
img = cv2.resize(bubu, (600, 600))

# Convert image from RGB to BGR
img1 = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# Export the image
cv2.imwrite(r'C:\Users\user\OneDrive\Desktop\buhari.jpg', img1)
```

```python
# Resize through %
width = 0.5
height = 0.5

bubu_r = cv2.resize(bubu, (0, 0), bubu, width, height)
plt.imshow(bubu_r)
```

```python
# Flipping the image (0 - Flips Horizonatal Axis)
bubu_f = cv2.flip(bubu, 0)
plt.imshow(bubu_f)
```

```python
bubu_d = cv2.flip(bubu, -1)
plt.imshow(bubu_d)
```

## Drawing on Images

```python
# Create a blank Canva
blank = np.zeros(shape = (600, 600, 3), dtype = np.int16)
plt.imshow(blank)
```

```python
# Draw a rectangle - pt1(Top Left Corner) pt2(Bottom Right Corner)
cv2.rectangle(blank, pt1 = (200, 200), pt2 = (400, 400), color = (250,0,0), thickness = 3)

plt.imshow(blank)
```

```python
cv2.rectangle(blank, pt1 = (250, 250), pt2 = (350, 350), color = (0, 250, 0), thickness = 3)

plt.imshow(blank)
```

```python
# Drawing a Circle
cv2.circle(blank, center = (300, 300), radius = 30, color = (0, 0, 250), thickness = 3)

plt.imshow(blank)
```

```python
# Drawing a Circle
cv2.circle(blank, center = (300, 300), radius = 200, color = (0, 0, 255), thickness = 5)

plt.imshow(blank)
```

```python
cv2.circle(blank, center = (500, 50), radius = 50, color = (0, 0, 255), thickness = -1)

plt.imshow(blank)
```

```python
# Create a blank Canva
blank = np.zeros(shape = (600, 600,3), dtype = np.int16)
plt.imshow(blank)
```

```python
# Draw a Traffic Light
cv2.rectangle(blank, pt1 = (150, 50), pt2 = (450, 550), color = (255,255, 255), thickness = 3)
plt.imshow(blank)
```

```python
cv2.circle(blank, center = (300, 150), radius = 70, color = (255,0, 0), thickness = -1)
cv2.circle(blank, center = (300, 300), radius = 70, color = (250,255, 1), thickness = -1)
cv2.circle(blank, center = (300, 450), radius = 70, color = (0,255, 0), thickness = -1)

plt.imshow(blank)
```
